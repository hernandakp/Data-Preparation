**4. DIMENSIONALITY REDUCTION**

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
pca = PCA(n_components=4)
pca.fit(df_numeric)
pca_samples = pca.transform(df_numeric)
ps = pd.DataFrame(pca_samples)
ps.head()
```

```python
tocluster = pd.DataFrame(ps[[0,2]])
print (tocluster.shape)
print (tocluster.head())
fig = plt.figure(figsize=(8,8))
plt.plot(tocluster[0], tocluster[2], 'o', markersize=2, color='maroon', alpha=0.5, label='class1')
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.legend()
plt.show()
```

```python
clusterer = KMeans(n_clusters=4,random_state=42).fit(tocluster)
centers = clusterer.cluster_centers_
c_preds = clusterer.predict(tocluster)
print(centers)
```

```python
fig = plt.figure(figsize=(8,8))
colors = ['orange','blue','purple','green']
colored = [colors[k] for k in c_preds]
plt.scatter(tocluster[0],tocluster[2],  color = colored)
for ci,c in enumerate(centers):
    plt.plot(c[0], c[1], '*', markersize=8, color='red', alpha=0.9, label=''+str(ci))
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.legend()
plt.show()
```
